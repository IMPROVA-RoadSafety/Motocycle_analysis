# Data processing

## Filters

```python
# Keep only rows where "MOTOCICLISTA" == 1
motocycle_df = decoded_df.copy() 
motocycle_df = motocycle_df[motocycle_df.iloc[:, 22] == 1]

print("After filtering motorcycle/moped users:", motocycle_df.shape)
```

```python
# General variables
general_columns = [motocycle_df.columns[12]] + motocycle_df.columns[15:21].tolist()  + [motocycle_df.columns[22]] + [motocycle_df.columns[26]]

part2_columns = motocycle_df.columns[29:127].tolist()

part7_columns = motocycle_df.columns[448:454].tolist()

print("General columns:", general_columns)
print("Part 2 columns:", part2_columns)
print("Part 7 columns:", part7_columns)

motocycle_df = motocycle_df[general_columns + part2_columns + part7_columns]
print("After selecting relevant columns:", motocycle_df.shape)
```

## Rider & context variables to describe


```python

# Create profile report
profile = ProfileReport(motocycle_df, title="All motocycle related variables for motocyclists", explorative=True)

# To save as an HTML file
profile.to_file(REPORTS_DIR/"only_motocycles.html")
```
